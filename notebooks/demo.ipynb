## 3. Check data is present

```python
# TODO (Issue #2): Download Flickr8k and place under data/flickr8k/
#   Images:   data/flickr8k/images/   (~1 GB)
#   Captions: data/flickr8k/captions.txt
#   Splits:   data/flickr8k/Flickr_8k.{train,dev,test}Images.txt
#
# Flickr8k is available via:
#   https://illinois.edu/fb/sec/1713398  (request form)
# or via Kaggle:
#   !kaggle datasets download -d adityajn105/flickr8k

import os
data_root = 'data/flickr8k'
required = [
    os.path.join(data_root, 'images'),
    os.path.join(data_root, 'captions.txt'),
    os.path.join(data_root, 'Flickr_8k.trainImages.txt'),
    os.path.join(data_root, 'Flickr_8k.devImages.txt'),
    os.path.join(data_root, 'Flickr_8k.testImages.txt'),
]
all_present = True
for path in required:
    exists = os.path.exists(path)
    status = '✓' if exists else '✗ MISSING'
    print(f'  {status}  {path}')
    if not exists:
        all_present = False

if all_present:
    n_images = len(os.listdir(os.path.join(data_root, 'images')))
    print(f'\nFound {n_images} images. Ready to proceed.')
else:
    print('\nPlease add the missing files before continuing.')
```
